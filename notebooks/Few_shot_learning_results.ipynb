# Generate Plots for Few Shot Learning

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
import pandas as pd 
```

```python
def main(result_path, test_size=-1):
    df = pd.read_csv(result_path, sep='\t')
    df["kshot"][df.loc[:, 'kshot'] == -1]= test_size
    sns.catplot(x="classifier_type", y="acc", hue="kshot", data=df, jitter=False,palette="ch:.25", size=5, aspect=1.5)
    sns.catplot(x="classifier_type", y="recall", hue="kshot", data=df, jitter=False,palette="ch:.25", size=5, aspect=1.5)
    sns.catplot(x="classifier_type", y="train_time", hue="kshot", data=df,palette="ch:.25", jitter=False, aspect=1.5)
    sns.catplot(x="classifier_type", y="test_time", hue="kshot", data=df,palette="ch:.25", size=5, aspect=1.5)
    
```
