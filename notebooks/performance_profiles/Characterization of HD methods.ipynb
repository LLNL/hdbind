```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
```

```python
def plot_encode_vs_time(df):
    f, ax = plt.subplots(1,2, figsize=(15,5), sharex=True)
    ax = ax.flatten()
    sns.scatterplot(df["encoding_dim"], df["train_encode_time"], ax=ax[0])
    ax[0].set(xscale="log", yscale="log")
    sns.scatterplot(df["encoding_dim"], df["test_encode_time"], ax=ax[1])
    ax[1].set(xscale="log", yscale="log")

    
def plot_encode_vs_metrics(df):
    f, ax = plt.subplots(2,1, figsize=(10,10), sharex=True)
    ax = ax.flatten()
#     sns.scatterplot(df["encoding_dim"], df["f1_score"], ax=ax[0],c=df["iterations"], cmap=sns.color_palette("flare", as_cmap=True)

    sns.scatterplot(df["encoding_dim"], df["f1_score"], ax=ax[0],c=df["iterations"])
    ax[0].set(xscale="log")
    sns.scatterplot(df["encoding_dim"], df["accuracy_score"], ax=ax[1], c=df["iterations"])
    ax[1].set(xscale="log")
    
    plt.legend()
    
def main(csv_path):
    
    df = pd.read_csv(csv_path, index_col=0)
    
    plot_encode_vs_time(df)
    
    plot_encode_vs_metrics(df)
    
```
