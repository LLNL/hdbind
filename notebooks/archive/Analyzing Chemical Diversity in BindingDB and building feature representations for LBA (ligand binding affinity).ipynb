```python
def load_binding_db_data():
#         bind_db_meta_df = PandasTools.LoadSDF(binding_db_sdf_path,smilesName='SMILES',molColName='Molecule',
#            includeFingerprints=False, strictParsing=False)
        bind_db_pwise_dist_df = pd.read_csv(binding_db_pairwise_dist_path, sep='\t')
        print(bind_db_pwise_dist_df.shape)
        
        bind_db_pwise_dist_part1_df = bind_db_pwise_dist_df[["ChEMBL1", "SMILES1"]]
#         bind_db_pwise_dist_part1_df = pd.merge(bind_db_pwise_dist_df[["ChEMBL1", "SMILES1"]],
#                                            bind_db_meta_df, left_on="SMILES1", right_on="SMILES")
        bind_db_pwise_dist_part1_df = bind_db_pwise_dist_part1_df.rename(columns={"SMILES1": "SMILES", "ChEMBL1": "ChEMBL_ID"})

        bind_db_pwise_dist_part2_df = bind_db_pwise_dist_df[["ChEMBL2", "SMILES2"]]
#         bind_db_pwise_dist_part2_df = pd.merge(bind_db_pwise_dist_df,
#                                            bind_db_meta_df, left_on="SMILES2", right_on="SMILES")
        bind_db_pwise_dist_part2_df = bind_db_pwise_dist_part2_df.rename(columns={"SMILES2": "SMILES","ChEMBL2": "ChEMBL_ID"})

        
        full_df = pd.concat([bind_db_pwise_dist_part1_df, bind_db_pwise_dist_part2_df])
        full_df = full_df.drop_duplicates()
        return full_df
        
#         return bind_db_pwise_dist_df
```

```python
bind_db_df = load_binding_db_data()
```

```python
def plot_pwise_dists(df):
    
    f, ax = plt.subplots(1,1, figsize=(10,7))
    sns.distplot(df['Tanimoto'], ax=ax)
    ax.set_title("bindingDB Ligand Tanimoto dist")
    plt.savefig("bindingdb_ligand_tanimoto_dist.png", dpi=600)
```
