```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
import pandas as pd
from pathlib import Path
from tqdm import tqdm
import numpy as np
from collections import defaultdict
```

```python
def get_dude_train_data_size_by_target(data_dir=Path("/usr/WS1/jones289/hd-cuda-master/datasets/dude/deepchem_feats"),
                                      feat_type="rdkit"):
    size_dict = defaultdict(int)
    for path in data_dir.glob(f"*/{feat_type}/train.npy"):
        target = path.parent.parent.name
        data = np.load(path)
        size_dict[target] = data.shape[0]
        
    return size_dict

def get_dude_test_data_size_by_target(data_dir=Path("/usr/WS1/jones289/hd-cuda-master/datasets/dude/deepchem_feats"),
                                     feat_type="rdkit"):
    size_dict = defaultdict(int)
    for path in data_dir.glob(f"*/{feat_type}/test.npy"):
        target = path.parent.parent.name
        data = np.load(path)
        size_dict[target] = data.shape[0]
        
    return size_dict

def get_dude_class_ratio_by_target(data_dir=Path("/usr/WS1/jones289/hd-cuda-master/datasets/dude/deepchem_feats"),
                                  feat_type="rdkit"):
    class_dict = defaultdict(int)
    for path in data_dir.glob(f"*/{feat_type}/train.npy"):
        target = path.parent.parent.name
        train_data = np.load(path)
#         test_data = np.load(path.with_name('test.npy'))
        
        class_dict[target] = (train_data[train_data[:, -1]== 1]).shape[0]/train_data.shape[0]
#         print(class_dict)
    return class_dict


def load_data(path):
    df = pd.read_csv(path, sep="\t")
    target = path.parent.parent.name
    print(target)
    train_size_dict = get_dude_train_data_size_by_target()
    test_size_dict = get_dude_test_data_size_by_target()

    for col in ["acc", "recall-micro", "recall-macro", "precision-micro", "precision-macro"]:
#         import pdb
#         pdb.set_trace()
        df[col] = df[col].apply(lambda x: float(x.replace("tensor(", "")[:-1]))
    
    df.loc[df["model"] != "HD", "encode_time"] =  0
    df["train+encode_time"] = df["train_time"]  + df["encode_time"]
    
    
#     = df["train_time"] + df["encode_time"]
    df[df["model"] == "HD"]["train+encode_time"].apply(lambda x: 0)
    
    df["train_size"] = [train_size_dict[target]] * df.shape[0]
    df["test_size"] =  [test_size_dict[target]] * df.shape[0]
    df["target"] = [target] * df.shape[0]
#     import pdb
#     pdb.set_trace()
    return df
    
    
    
def main(path, size_dict, class_dict):
    df = load_data(path)
    
    target = path.parent.parent.name
    feat = path.parent.name
    
    
    
    
    f, ax = plt.subplots(1,5, figsize=(25,7))
    f.suptitle(f"{target}-{feat}-{size_dict[target]}-{class_dict[target]:0.2f}")
#     sns.boxplot(x="model", y="acc", data=df, ax=ax[0])
#     sns.boxplot(x="model", y="recall-macro", data=df, ax=ax[1])
#     sns.boxplot(x="model", y="train_time", data=df[df["model"] != "RANDOM"], ax=ax[2])
#     sns.boxplot(x="model", y="train+encode_time", data=df[df["model"] != "RANDOM"], ax=ax[3])
#     sns.boxplot(x="model", y="test_time", data=df[df["model"] != "RANDOM"], ax=ax[4])
    sns.violinplot(x="model", y="acc", data=df, ax=ax[0], cut=0)
    sns.violinplot(x="model", y="recall-macro", data=df, ax=ax[1], cut=0)
    sns.violinplot(x="model", y="train_time", data=df[df["model"] != "RANDOM"], ax=ax[2], cut=0)
    sns.violinplot(x="model", y="train+encode_time", data=df[df["model"] != "RANDOM"], ax=ax[3], cut=0)
    sns.violinplot(x="model", y="test_time", data=df[df["model"] != "RANDOM"], ax=ax[4], cut=0)
```

```python
def make_time_vs_data_size_plot():
    data_dir = Path("/usr/WS1/jones289/hd-cuda-master/datasets/dude/deepchem_feats")

    size_dict = get_dude_train_data_size_by_target()
    class_dict = get_dude_class_ratio_by_target()
    
    df_list = []
    for path in data_dir.glob("*/rdkit/results.csv"):
#         print(path)
#         main(path, size_dict, class_dict)
        df = load_data(path)
        df_list.append(df)
    
    full_df = pd.concat(df_list)
    
    f, ax = plt.subplots(1,1, figsize=(12,6))
    sns.scatterplot(full_df[full_df["model"] == "HD"].groupby("target")["train_size"].mean().values,
                full_df[full_df["model"] == "HD"].groupby("target")["train_time"].mean().values, ax=ax)
    sns.scatterplot(full_df[full_df["model"] == "RF"].groupby("target")["train_size"].mean().values,
                full_df[full_df["model"] == "RF"].groupby("target")["train_time"].mean().values)
    ax.set(yscale="log")

    
    return full_df
```

```python
full_df = make_time_vs_data_size_plot()
```

```python
def make_metric_vs_train_size_plot(metric):
    data_dir = Path("/usr/WS1/jones289/hd-cuda-master/datasets/dude/deepchem_feats")

    size_dict = get_dude_train_data_size_by_target()
    class_dict = get_dude_class_ratio_by_target()
    
    df_list = []
    for path in data_dir.glob("*/rdkit/results.csv"):
#         print(path)
#         main(path, size_dict, class_dict)
        df = load_data(path)
        df_list.append(df)
    
    full_df = pd.concat(df_list)
    
    f, ax = plt.subplots(1,1, figsize=(12,6))
    sns.scatterplot(full_df[full_df["model"] == "HD"].groupby("target")["train_size"].mean().values,
                full_df[full_df["model"] == "HD"].groupby("target")[metric].mean().values, ax=ax)
    sns.scatterplot(full_df[full_df["model"] == "RF"].groupby("target")["train_size"].mean().values,
                full_df[full_df["model"] == "RF"].groupby("target")[metric].mean().values)
#     ax.set(yscale="log")

    
    return full_df
```
