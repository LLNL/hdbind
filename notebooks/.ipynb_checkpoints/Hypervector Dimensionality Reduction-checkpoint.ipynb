```python
import torch
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import math
import matplotlib.pyplot as plt 
import seaborn as sns

import sys
# sys.path.append('..')
sys.path.insert(0, "/g/g13/jones289/workspace/hd-cuda-master/hdpy")
```

```python
# hv_dir = "/g/g13/jones289/workspace/hd-cuda-master/hdpy/hdpy/.hd_cache"
hv_dir = Path("/g/g13/jones289/workspace/hd-cuda-master/hdpy/hdpy/.hd_cache/125")
result_dir = Path("/g/g13/jones289/workspace/hd-cuda-master/hdpy/hdpy/results/125")



color_dict = {0: "red",
           1: "blue"}

size_dict = {0: 10,
           1: 30}

alpha_dict = {0: 0.25,
             1: 1.0}

marker_dict = {0: "o",
              1: "X"}

name_dict = {0: "inactive",
            1: "active"}

plot_dict = {"color": color_dict,
          "size": size_dict,
            "alpha": alpha_dict,
            "marker": marker_dict,
            "name": name_dict}

import pickle

def make_ax_scatter_plot(data,dataset, datatype, labels, ax, target:str):

    comps = None
    
    pca_path = Path(f"{dataset}-{target}-{datatype}-pca.pkl")

    if not pca_path.exists():

        pca_pipe = make_pipeline(StandardScaler(), PCA())
        pca_pipe.fit(data)
        comps = pca_pipe.transform(data)
        
        with open(pca_path, 'wb') as handle:
            pickle.dump({"pca_pipe":pca_pipe, "comps":comps }, handle)
    
    else:
        with open(pca_path, "rb") as handle:
            
            pca_dict = pickle.load(handle)
            comps = pca_dict["comps"]

    
    pca_df = pd.DataFrame({"x": (comps[:, 0]).squeeze(), "y": (comps[:, 1]).squeeze(), "label": labels})
        
#     sns.scatterplot(data=pca_df, x="x", y="y", hue="label", ax=ax)
    
    
    line_list = []
    for i in range(2):
        
        class_comps= comps[labels == i, :]
        class_labels = labels[labels == i]
        
        sizes = [plot_dict["size"][x] for x in class_labels]
        alpha = [plot_dict["alpha"][x] for x in class_labels]
        marker = [plot_dict["marker"][x] for x in class_labels]
        
        
        plot_i = sns.scatterplot(x=class_comps[:, 0], y=class_comps[:, 1], ax=ax, 
                                 color=plot_dict["color"][i], 
                                  s=sizes, 
                                 alpha=alpha, 
                                 style=class_labels, 
                                 markers=plot_dict["marker"], 
#                                  marker=
                                 legend="full", 
                                 zorder=i)
        
#         extract legend handles and append to line_list
        handles, _ = plot_i.get_legend_handles_labels()
        line_list.append(handles[i])
        
#     fig = ax.get_figure()
    ax.legend(line_list, [f"{target}-{plot_dict['name'][0]}", f"{target}-{plot_dict['name'][1]}"], loc='upper center', ncol=2)
    leg = ax.get_legend()
    leg.legendHandles[0].set_color(color_dict[0])
    leg.legendHandles[1].set_color(color_dict[1])






def main(model:str, dataset:str, split:str, tokenizer:str, ngram_order:int,
         dev_mode:bool, figsize:tuple, nrows:int, ncols:int,
        datatype:str):
    
    # 'datatype' should be either "hv" or "ecfp"
    
    
    path_list = list(hv_dir.glob(f"{model}/{dataset}/{split}/*"))
    
    if dev_mode:
        path_list=path_list[0:4]
    
#     fig_size = int(math.sqrt(len(path_list))) + 1

#     f, ax = plt.subplots(nrows=fig_size,ncols=fig_size, figsize=figsize, sharex=True, sharey=True)
    f, ax = plt.subplots(nrows=nrows,ncols=ncols, figsize=figsize, sharex=True, sharey=True)
    ax = ax.flatten()
    
    
    
    for i, path in enumerate(path_list):
#         print(path)
        
        target = path.name
        
        test_labels = np.load(result_dir / Path(f"{dataset}_{split}_{target}_labels_test.npy"))

        data = None
        if datatype == "hv":
            test_hvs = torch.load(path / Path("test_dataset_hv.pth")).numpy()
            data = test_hvs
        elif datatype == "ecfp":
            
            result_pkl_path = None
            if dataset == "lit-pcba":
                result_pkl_path = Path(f"{dataset.replace('-', '_')}.{target}.{model}.{tokenizer}.{ngram_order}.pkl")
            elif dataset == "dude":
                result_pkl_path = Path(f"{dataset.replace('-', '_')}.{split}.{target}.{model}.{tokenizer}.{ngram_order}.pkl")
            
            
            with open(result_dir / result_pkl_path, "rb") as handle:
                result_pkl = pickle.load(handle)
                ecfps = result_pkl["x_test"]
                data = ecfps
        
        print(sum(test_labels))
        
#         print(test_hvs, test_hvs.exists(), test_labels, test_labels.exists())
#         print(test_hvs.shape , test_labels.shape)
        
        make_ax_scatter_plot(dataset=dataset, datatype=datatype, data=data, labels=test_labels.squeeze(), ax=ax[i], target=target)
        

    plt.tight_layout()
```

## Dev mode

```python
main(model="ecfp", dataset="lit-pcba", datatype="ecfp", split="random", tokenizer="atomwise", ngram_order=0, dev_mode=False, figsize=(15,15), nrows=5, ncols=3)
plt.savefig("lit-pcba-ecfp-scatter.png", dpi=600)
```
