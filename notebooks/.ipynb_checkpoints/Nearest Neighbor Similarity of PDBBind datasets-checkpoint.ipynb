```python
from atomsci.ddm.pipeline import chem_diversity as cd
from atomsci.ddm.utils.struct_utils import base_smiles_from_smiles

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_smiles(smiles_file):
    smiles_df = pd.read_csv(smiles_file,header=None, sep='\t')
    smiles_df['pdbid'] = smiles_df[1].apply(lambda x: x.split('_')[0])
    smiles_df['smiles'] = smiles_df[0]
    smiles_df['rdkit_smiles'] = smiles_df['smiles'].apply(base_smiles_from_smiles)

    return smiles_df[['pdbid', 'smiles', 'rdkit_smiles']]
smiles_df = load_smiles("/p/lustre2/jones289/data/raw_data/v2019/v2019.smi")
```

```python
smiles_df
```

```python
def main(smiles_df):
    
    pdbspheres_id_df = pd.read_csv("/p/lustre2/jones289/data/pdbbind/metadata/PDBSPHERES_EVAL_SET.csv", header=None)
    pdbspheres_id_df['pdbid'] = pdbspheres_id_df[0]
    
    # This is test set
    pdbbind_2019_df = pd.read_csv("/p/lustre2/jones289/data/pdbbind/metadata/pdbbind_v2019_metadata.csv")
    pdbbind_2019_df = pdbbind_2019_df[pdbbind_2019_df["bind"].apply(lambda x: x!=2)]
    pdbspheres_smiles_df = pd.merge(smiles_df, pdbspheres_id_df, on='pdbid')
    pdbspheres_smiles_df = pd.merge(pdbspheres_smiles_df, pdbbind_2019_df, on='pdbid')
    
    pdbbind_2016_df = pd.read_csv("/p/lustre2/jones289/data/pdbbind/metadata/pdbbind_2016_train_val_test.csv")
    pdbbind_2016_df = pdbbind_2016_df[pdbbind_2016_df["bind"].apply(lambda x: x!=2)]
    pdbbind_2016_smiles_df = pd.merge(smiles_df, pdbbind_2016_df, on='pdbid')
    
    core_set_smiles_df = pdbbind_2016_smiles_df[pdbbind_2016_smiles_df['set'] == 'core_test']
    refined_smiles_df = pdbbind_2016_smiles_df[pdbbind_2016_smiles_df['set'].apply(lambda x: "refined" in x)]
    
    refined_core_dists = cd.calc_dist_smiles('ecfp', 'tanimoto', core_set_smiles_df['rdkit_smiles'].values.tolist(), 
                                refined_smiles_df['rdkit_smiles'].values.tolist(), calc_type='nearest', num_nearest=1)
    
    refined_pdbspheres_dists = cd.calc_dist_smiles('ecfp', 'tanimoto', pdbspheres_smiles_df['rdkit_smiles'].values.tolist(), 
                                refined_smiles_df['rdkit_smiles'].values.tolist(), calc_type='nearest', num_nearest=1)
    
    
    
    
    
    print("refined (train): ", refined_smiles_df.shape[0])
    print("core (test): ",core_set_smiles_df.shape[0])
    print("pdbspheres (test): ", pdbspheres_smiles_df.shape[0])
    
    f, ax = plt.subplots(1,1, figsize=(12, 7), sharex=True, sharey=True)
    ax.set_title("Tanimoto Distance from NN in Training Set")
    sns.distplot(refined_core_dists, ax=ax, label="core 2016")
#     ax[0].set_title(f"core 2016 (test) distance to NN in refined (train)")

    sns.distplot(refined_pdbspheres_dists, ax=ax, label="pdbspheres 2019")
#     ax[1].set_title(f"pdbspheres (test) distance to NN in refined (train)")
    plt.legend()
    
    
main(smiles_df)
```
